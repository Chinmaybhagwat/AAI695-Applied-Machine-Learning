```python
# Given code for polynomial function 5x + 15x**2 + 2x***3
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

```python
# Quastion-1: Please plot the noisy data and the polynomial you found (in the same figure).
#You can use any value of m selected from 2, 3, 4, 5, 6.

x = x.ravel() # to make contiguous flattened array
y = y.ravel()
z = np.polyfit(x, y, 2) # choosing m =2
p = np.poly1d(z) # to make z a polynomial  function
plt.plot(x,y, 'ro')
plt.plot(x, p(x),'yo', label = 'm = 2')
plt.legend()
plt.show
```

```python
# Question-2: Plot MSE versus order m, for m = 1, 2, 3, 4, 5, 6, 7, 8 respectively. Identify the best choice of m.

x = x.ravel() # to make contiguous flattened array
y = y.ravel()
m = []
for i in [1,2,3,4,5,6,7,8]:
    z = np.polyfit(x,y,i)
    p = np.poly1d(z)
    m.append(((sum((x-p(x))**2)))/number_of_samples) # adding MSE formula
    plt.title("M value 1 to 8")
    plt.plot(x,y, 'ro')
    plt.plot(x,p(x), 'yo', label = 'polynomial value')
    plt.legend()
    plt.show()
print("MSE: ",m)
plt.plot([1,2,3,4,5,6,7,8],m)
plt.xlabel('M')
plt.ylabel('MSE')
plt.show()  
```

```python
# Question-3: Change variable noise_scale to 200, 300, 400, 600, 800, 1000 respectively, re-run the algorithm
#and plot the polynomials with the m found in 2). Discuss the impact of noise scale to the $
#accuracy of the returned parameters.

for noise_scale in [200, 300, 400, 600, 800, 1000]:
    number_of_samples = 50
    x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    x = x.ravel()
    y = y.ravel()
    z = np.polyfit(x, y, 3) # from question 2 we found that 3 is the best choice for m
    p = np.poly1d(z)
    plt.title(f'noise_scale = {noise_scale}')
    plt.plot(x, y, 'ro')
    plt.plot(x,p(x),'yo',label = 'Polynomial value')
    plt.legend()
    plt.show()
    
```

```python
# Question-4: Change variable number_of_samples to 40, 30, 20, 10 respectively, re-ran the algorithm and plot
#the polynomials with the m found in 2). Discuss the impact of the number of samples to the 
#accuracy of the returned parameters.

for number_of_samples in [40, 30, 20, 10]:
    noise_scale = 100
    x = 25*(np.random.rand(number_of_samples, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    x = x.ravel()
    y = y.ravel()
    z = np.polyfit(x, y, 3) # from question 2 we found that 3 is the best choice for m
    p = np.poly1d(z)
    plt.title(f'number_of_samples = {number_of_samples}')
    plt.plot(x, y, 'ro')
    plt.plot(x,p(x),'yo',label = 'Polynomial value')
    plt.legend()
    plt.show()
```
